```python
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv("movies.csv")
```

```python
data.describe()
```

```python
data.isna().sum()
```

```python
data.head()
```

```python
data.drop('index', axis=1,inplace=True)
```

```python
data.head(5)
```

```python
data['year_of_release']= data['year_of_release'].abs()
```

```python
data.head(5)
```

```python
data['run_time'] = data['run_time'].str.replace(' min','')
```

```python
data.head(5)
```

```python
data['run_time'] = data['run_time'].astype(float)
```

```python
data.info()
```

```python
data['gross_total'] = data['gross_total'].str.strip('$M')
```

```python
data.head(5)
```

```python
data['gross_total'] = data['gross_total'].astype(float)
```

```python
data.rename(columns={'gross_total': 'gross_total_m'}, inplace=True)
```

```python
data['main_genre'] = data['genre'].str.split(',').str[0]
```

```python
data.head()
```

```python
import matplotlib.pyplot as plt

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(18,5))

top5 = data.sort_values(by='imdb_rating', ascending=False).head(5)
sns.barplot(data=top5, x='imdb_rating', y='movie_name', ax=ax1, color = 'grey', edgecolor='#ff7f7f', linewidth=2)
ax1.set_title("Top 5 Movies")

avg_rating = data.groupby('year_of_release')['imdb_rating'].mean().reset_index()
sns.lineplot(data=avg_rating, x='year_of_release', y='imdb_rating', marker='o', ax=ax2, color='black')
ax2.lines[0].set_markerfacecolor('white')
ax2.lines[0].set_markeredgecolor('red')
ax2.set_title("Rating by Year")



genre_avg = data.groupby('main_genre')['imdb_rating'].mean().sort_values(ascending=False)
sns.scatterplot(x=genre_avg.values, y=genre_avg.index, ax=ax3, color='red', edgecolor='grey')
ax3.set_title("Rating by Genre")

plt.tight_layout()
plt.show()
```
